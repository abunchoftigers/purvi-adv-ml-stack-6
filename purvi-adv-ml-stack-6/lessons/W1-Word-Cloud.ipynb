```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS

import os
import re
```

```python
STOPWORDS
```

data

```python
# Define the filepath
fpath= "../data/Alice_in_Wonderland.txt"
# Use with open syntax
with open(fpath) as f:
    txt = f.read()
# Report length of the text
print(f"There are {len(txt)} characters in the full text.\n")

```

```python
# Preview first 1,000 characters (printing)
print(txt[:1000])
```

make a word cloud

```python
# Create an instance of a WordCloud and use the generate method
cloud = WordCloud(random_state = 123).generate(txt)
plt.imshow(cloud);
plt.axis('off');
```

define stopwords

```python
custom_stopwords = ['said', 'Alice', 's', *STOPWORDS]
```

word cloud with stopwords

```python
# Create an instance of a WordCloud and use the generate method
cloud = WordCloud(random_state=123, stopwords=custom_stopwords).generate(txt)
plt.imshow(cloud);
plt.axis('off');
```

change color palette

```python
# Use a different color map
cloud = WordCloud(
    random_state=123,
    stopwords = custom_stopwords,
    colormap="plasma"
).generate(txt)
plt.imshow(cloud)
plt.axis("off");

```

create an instance of a word cloud using .generate()

```python
cloud = WordCloud(
  random_state=123,
  stopwords = custom_stopwords,
  background_color='white',
  colormap='plasma',
  width=500,
  height=500,
  max_words=200,
  min_word_length=2,
).generate(txt)

plt.figure(figsize=(10,10))
plt.imshow(cloud)
plt.axis('off');
```
